```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here

#H0: Pokemon of type Dragon do not have, on average, more HP stats than Pokemon of type Grass.
#H1: Pokemon of type Dragon have, on average, more HP stats than Pokemon of type Grass.


#OR

#H0: mu dragon_HP <= mu grass_HP
#H1: mu dragon_HP > mu grass_HP


alpha = 0.05

dragon_hp = df[df["Type 1"] == "Dragon"]["HP"]
grass_hp = df[df["Type 1"] == "Grass"]["HP"]

#Using greter makes it a one sided test
st.ttest_ind(dragon_hp, grass_hp, alternative='greater')
```

```python
#Since the p value is lower than the significance level we reject the null hypothesis. Which means that in average Dragon type pokemons have more HP stats than Grass type pokemons.
```

```python
dragon_hp.mean()
grass_hp.mean()

print("Dragon HP mean:", dragon_hp.mean())
print("Grass HP mean:", grass_hp.mean())
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here

#H0: Legendary Pokemon have the same stats as non-Legendary Pokemon.
#H1: Legendary Pokemon have different stats than non-Legendary Pokemon

#OR

#H0: mu legendary_stats = mu non_legendary_stats
#H1: mu legendary_stats != mu non_legendary_stats

alpha = 0.05

stats = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]

legendary = df[df["Legendary"] == True]
non_legendary = df[df["Legendary"] == False]

for stat in stats:
   st.ttest_ind(legendary[stat], non_legendary[stat])
   print(f"Stat: {stat} - p-value: {st.ttest_ind(legendary[stat], non_legendary[stat])}")
```

```python
#Because all p values are smaller than the significance level, we reject the null hypothesis. This means that Legendary Pokémon have different stats than non-Legendary Pokémon
```
